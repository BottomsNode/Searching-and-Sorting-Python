# <center>Selection Sort Alogorithm</center>

```python
import tkinter as tk
from tkinter import messagebox
import time

class SelectionSortVisualizer:
    def __init__(self, master):
        self.master = master
        master.title("Selection Sort Visualization")
        master.geometry("600x450")  # Increased height for the label

        self.canvas = tk.Canvas(master, width=500, height=200)
        self.canvas.pack(pady=20)

        self.entry = tk.Entry(master)
        self.entry.pack(pady=5)
        self.entry.insert(0, "5 8 4 9 3 2 7 1 6")  # Default list for sorting

        self.start_button = tk.Button(master, text="Start Selection Sort", command=self.animate_selection_sort)
        self.start_button.pack(pady=5)

        self.rectangles = []
        self.steps_label = tk.Label(master, text="", font=("Arial", 12))  # Label to display steps
        self.steps_label.pack()

    def animate_selection_sort(self):
        try:
            nums = [int(num) for num in self.entry.get().split()]
        except ValueError:
            messagebox.showerror("Error", "Invalid input. Please enter integers separated by spaces.")
            return

        self.steps_label.config(text="Starting Selection Sort...")
        self.draw_rectangles(nums)
        self.selection_sort(nums)
        self.steps_label.config(text="Selection Sort Completed. List is now sorted!")

    def selection_sort(self, nums):
        n = len(nums)
        for i in range(n):
            min_idx = i
            for j in range(i+1, n):
                if nums[j] < nums[min_idx]:
                    min_idx = j
            nums[i], nums[min_idx] = nums[min_idx], nums[i]
            self.update_canvas(nums, i, min_idx)
            self.master.update()  # Update the window to show changes
            time.sleep(0.6)  # Add animation delay
            self.steps_label.config(text=f"Swapped {nums[i]} with {nums[min_idx]}")  # Update explanation
            self.master.update()  # Update the window to show explanation

    def draw_rectangles(self, nums):
        self.canvas.delete("rectangles")
        self.rectangles.clear()

        for i, num in enumerate(nums):
            x0, y0 = 20 + i * 50, 100
            x1, y1 = x0 + 40, 150
            rect = self.canvas.create_rectangle(x0, y0, x1, y1, fill="white", tags="rectangles")
            self.rectangles.append(rect)
            self.canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(num), tags="rectangles")

    def update_canvas(self, nums, i, min_idx):
        self.canvas.delete("rectangles")
        for j, num in enumerate(nums):
            color = "green" if j == i or j == min_idx else "white"
            x0, y0 = 20 + j * 50, 100
            x1, y1 = x0 + 40, 150
            rect = self.canvas.create_rectangle(x0, y0, x1, y1, fill=color, tags="rectangles")
            self.rectangles.append(rect)
            self.canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(num), tags="rectangles")

if __name__ == "__main__":
    root = tk.Tk()
    app = SelectionSortVisualizer(root)
    root.mainloop()
```
