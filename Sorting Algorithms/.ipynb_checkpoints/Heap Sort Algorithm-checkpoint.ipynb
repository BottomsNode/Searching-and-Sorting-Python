# <center>Heap Sort Alogorithm</center>

```python
import tkinter as tk
import time

class HeapSortVisualizer:
    def __init__(self, master):
        self.master = master
        master.title("Heap Sort Visualization")
        master.geometry("600x400")

        self.canvas = tk.Canvas(master, width=500, height=200)
        self.canvas.pack(pady=20)

        self.entry = tk.Entry(master)
        self.entry.pack(pady=5)
        self.entry.insert(0, "5 8 4 9 3 2 7 1 6")  # Default list for sorting

        self.start_button = tk.Button(master, text="Start Heap Sort", command=self.animate_heap_sort)
        self.start_button.pack(pady=5)

        self.steps_label = tk.Label(master, text="", font=("Arial", 12))
        self.steps_label.pack(pady=5)

        self.rectangles = []

    def animate_heap_sort(self):
        try:
            nums = [int(num) for num in self.entry.get().split()]
        except ValueError:
            messagebox.showerror("Error", "Invalid input. Please enter integers separated by spaces.")
            return

        self.steps_label.config(text="Starting Heap Sort...")
        self.draw_rectangles(nums)
        self.heap_sort(nums)

    def heapify(self, nums, n, i):
        largest = i
        l = 2 * i + 1
        r = 2 * i + 2

        if l < n and nums[l] > nums[largest]:
            largest = l

        if r < n and nums[r] > nums[largest]:
            largest = r

        if largest != i:
            nums[i], nums[largest] = nums[largest], nums[i]
            self.update_canvas(nums, i, largest, f"Swapped {nums[i]} with {nums[largest]}")
            time.sleep(0.5)
            self.heapify(nums, n, largest)

    def heap_sort(self, nums):
        n = len(nums)

        for i in range(n // 2 - 1, -1, -1):
            self.heapify(nums, n, i)

        for i in range(n - 1, 0, -1):
            nums[i], nums[0] = nums[0], nums[i]
            self.update_canvas(nums, i, 0, f"Swapped {nums[i]} with {nums[0]}")
            time.sleep(0.5)
            self.heapify(nums, i, 0)

        self.steps_label.config(text="Heap Sort Completed.")

    def draw_rectangles(self, nums):
        self.canvas.delete("rectangles")
        self.rectangles.clear()

        for i, num in enumerate(nums):
            x0, y0 = 20 + i * 50, 100
            x1, y1 = x0 + 40, 150
            rect = self.canvas.create_rectangle(x0, y0, x1, y1, fill="white", tags="rectangles")
            self.rectangles.append(rect)
            self.canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(num), tags="rectangles")

    def update_canvas(self, nums, i, j, explanation):
        self.canvas.delete("rectangles")
        for k, num in enumerate(nums):
            color = "green" if k == i or k == j else "white"
            x0, y0 = 20 + k * 50, 100
            x1, y1 = x0 + 40, 150
            rect = self.canvas.create_rectangle(x0, y0, x1, y1, fill=color, tags="rectangles")
            self.rectangles.append(rect)
            self.canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(num), tags="rectangles")
        self.steps_label.config(text=explanation)
        self.master.update()  # Update the canvas immediately
        time.sleep(1)  # Add a delay for better visualization

if __name__ == "__main__":
    root = tk.Tk()
    app = HeapSortVisualizer(root)
    root.mainloop()
```
