# <center>Quick Sort Alogorithm</center>

```python
import tkinter as tk
from tkinter import messagebox
import time

class QuickSortVisualizer:
    def __init__(self, master):
        self.master = master
        master.title("Quick Sort Visualization")
        master.geometry("600x450")  # Increased height for the label

        self.canvas = tk.Canvas(master, width=500, height=200)
        self.canvas.pack(pady=20)

        self.entry = tk.Entry(master)
        self.entry.pack(pady=5)
        self.entry.insert(0, "5 8 4 9 3 2 7 1 6")  # Default list for sorting

        self.start_button = tk.Button(master, text="Start Quick Sort", command=self.animate_quick_sort)
        self.start_button.pack(pady=5)

        self.rectangles = []
        self.steps_label = tk.Label(master, text="", font=("Arial", 12))  # Label to display steps
        self.steps_label.pack()

    def animate_quick_sort(self):
        try:
            nums = [int(num) for num in self.entry.get().split()]
        except ValueError:
            messagebox.showerror("Error", "Invalid input. Please enter integers separated by spaces.")
            return

        self.steps_label.config(text="Starting Quick Sort...")
        self.draw_rectangles(nums)
        self.quick_sort(nums, 0, len(nums) - 1)
        self.steps_label.config(text="Quick Sort Completed. List is now sorted!")

    def quick_sort(self, nums, low, high):
        if low < high:
            pi = self.partition(nums, low, high)
            self.quick_sort(nums, low, pi - 1)
            self.quick_sort(nums, pi + 1, high)

    def partition(self, nums, low, high):
        pivot = nums[high]
        i = low - 1

        for j in range(low, high):
            if nums[j] <= pivot:
                i += 1
                nums[i], nums[j] = nums[j], nums[i]
                self.update_canvas(nums, i, j)
                self.master.update()  # Update the window to show changes
                time.sleep(0.2)  # Add animation delay

        nums[i + 1], nums[high] = nums[high], nums[i + 1]
        self.update_canvas(nums, i + 1, high)
        self.master.update()  # Update the window to show changes
        time.sleep(0.2)  # Add animation delay
        return i + 1

    def draw_rectangles(self, nums):
        self.canvas.delete("rectangles")
        self.rectangles.clear()

        for i, num in enumerate(nums):
            x0, y0 = 20 + i * 50, 100
            x1, y1 = x0 + 40, 150
            rect = self.canvas.create_rectangle(x0, y0, x1, y1, fill="white", tags="rectangles")
            self.rectangles.append(rect)
            self.canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(num), tags="rectangles")

    def update_canvas(self, nums, i, j):
        self.canvas.delete("rectangles")
        for k, num in enumerate(nums):
            color = "green" if k == i or k == j else "white"
            x0, y0 = 20 + k * 50, 100
            x1, y1 = x0 + 40, 150
            rect = self.canvas.create_rectangle(x0, y0, x1, y1, fill=color, tags="rectangles")
            self.rectangles.append(rect)
            self.canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(num), tags="rectangles")

if __name__ == "__main__":
    root = tk.Tk()
    app = QuickSortVisualizer(root)
    root.mainloop()
```
