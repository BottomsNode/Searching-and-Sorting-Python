# <center>DFS Seach Alogorithm</center>

```python
import tkinter as tk
import time

def dfs_search(start, target, graph):
    """
    Perform DFS to find the target element in the graph.
    """
    stack = [start]
    visited = set()
    
    while stack:
        current = stack.pop()
        yield current  # Yield the current node being checked
        
        if current == target:
            return current
        
        visited.add(current)
        for neighbor in graph[current]:
            if neighbor not in visited:
                stack.append(neighbor)
    
    return -1

def animate_dfs():
    """
    Perform DFS animation.
    """
    graph = {0: [1, 2], 1: [3, 4], 2: [5, 6], 3: [], 4: [], 5: [], 6: []}
    start = 0
    target = int(entry_target.get())
    
    for node in dfs_search(start, target, graph):
        canvas.itemconfig(rectangles[node], fill="green")
        canvas.update()
        time.sleep(0.5)
        canvas.itemconfig(rectangles[node], fill="white")
        
        if node == target:
            result_label.config(text=f"Node {target} found!")
            return
    
    result_label.config(text="Node not found!")

# Create the main window
window = tk.Tk()
window.geometry("600x350")
window.title("DFS Visualization - By Nishit Shivdasani")

# Create canvas to display the graph nodes
canvas = tk.Canvas(window, width=500, height=200)
canvas.pack(pady=20)

# Entry for the target node
entry_target = tk.Entry(window)
entry_target.pack(pady=5)
entry_target.insert(0, "5")  # Default target node

# Button to start animation
dfs_button = tk.Button(window, text="Start DFS", command=animate_dfs)
dfs_button.pack(pady=5)

# Label to display search result
result_label = tk.Label(window, text="", font=('Calibri', 12))
result_label.pack(pady=10)

# Create rectangles for visualization
rectangles = []
for i in range(7):
    x0, y0 = 50 + i * 60, 50
    x1, y1 = x0 + 40, 90
    rectangles.append(canvas.create_rectangle(x0, y0, x1, y1, fill="white"))
    canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=i)

window.mainloop()
```
